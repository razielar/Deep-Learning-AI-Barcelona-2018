### Libraries:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd

%matplotlib inline
sns.set_style('white')
```

```python
x = np.linspace(start= -10, stop= 10, num= 1000)
y = stats.norm.pdf(x, loc=0, scale=1.5) #mean 0, sd = 1.5
plt.plot(x, y)
```
